## 活性化関数

以下でそれぞれの活性化関数を実装しグラフを描画することで、それぞれの特徴をとらえる

※ ニューラルネットワークでは **非線形** の活性化関数しか使用できない。。が、完全にはなぜが理解できていない


```python
# import

import numpy as np
from matplotlib import pyplot as plt
```

### ステップ関数

入力が閾値を超えた段階で1を返し、それ以外の場合は0を返すような関数


```python
def step_function(x: np.ndarray) -> np.ndarray:
    return np.array(x > 0, dtype=int)


x = np.arange(-5, 5, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.2, 1.2)
plt.show()
```

### シグモイド関数

ステップ関数に比べるとすごいなめらか

#### 定義

$$ h(x) = {1 \over 1 + \exp(-x)} $$



```python
def sigmoid(x: np.ndarray) -> np.ndarray:
    return 1 / (1 + np.exp(-x))


y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.2, 1.2)
plt.show()
```

```python
# シグモイドとステップ関数の比較

y_step = step_function(x)
y_sig = sigmoid(x)

plt.plot(x, y_step, label='step', linestyle='--')
plt.plot(x, y_sig, label='sigmoid')
plt.ylim(-0.2, 1.2)
plt.show()
```

### ReLU(Rectified Linear Unit)関数

#### 定義

$$
h(x) = \begin{cases}
    x (x > 0) \\
    0 (x \leq 0)
\end{cases}
$$

```python
def relu(x: np.ndarray) -> np.ndarray:
    return np.maximum(0, x)

y = relu(x)
plt.plot(x, y)
plt.ylim(-0.1, 5.1)
plt.show()
```

### ソフトマックス関数

正確には、出力層での分類問題の出力のための変換として使われる関数

#### 定義

$$ y_k = { \exp(a_k) \over \sum_{i=1}^{n} \exp(a_i) $$


```python
def softmax(a: np.ndarray) -> np.ndarray:
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    return exp_a / sum_exp_a

x_1 = np.arange(-5, 100, 0.1)
y = softmax(x_1)
plt.plot(x_1, y)
plt.show()
```
