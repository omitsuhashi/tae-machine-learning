## 活性化関数

ニューラルネットワークでは **非線形** の活性化関数しか使用できない



---

以下でそれぞれの活性化関数を実装しグラフを描画することで、それぞれの特徴をとらえる

```python
# import

import numpy as np
from matplotlib import pyplot as plt
```

### ステップ関数

入力が閾値を超えた段階で1を返し、それ以外の場合は0を返すような関数


```python
def step_function(x: np.ndarray) -> np.ndarray:
    return np.array(x > 0, dtype=int)


x = np.arange(-5, 5, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.2, 1.2)
plt.show()
```

### シグモイド関数

ステップ関数に比べるとすごいなめらか

#### 定義

$$ h(x) = 1 \over 1 + \exp(-x) $$



```python
def sigmoid(x: np.ndarray) -> np.ndarray:
    return 1 / (1 + np.exp(-x))


y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.2, 1.2)
plt.show()
```

```python
# シグモイドとステップ関数の比較

y_step = step_function(x)
y_sig = sigmoid(x)

plt.plot(x, y_step, label='step', linestyle='--')
plt.plot(x, y_sig, label='sigmoid')
plt.ylim(-0.2, 1.2)
plt.show()
```
